```python
from langchain_community.document_loaders import PyPDFLoader
pdf = PyPDFLoader("ai_agents.pdf") 
document = pdf.load() 
print(document)
```

## Converting data to chunks

```python
# Transforming the loaded document into chunks
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
doc = text_splitter.split_documents(document)  
print(doc[:5])   
```

## Transforming text to vector for storing

```python
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma

embedding = OpenAIEmbeddings(api_key=OPEN_AI_API_KEY)
db = Chroma.from_documents(doc[:20], embedding)
```

## Retrieving the data from vector store for answering the query

```python
# Retrieving the most similar documents to a query from the vector store
query = "What is an AI agent?"
results= db.similarity_search(query)
raw_text= results[0].page_content
cleaned_text= "".join(raw_text.split("\n"))
cleaned_text
```
